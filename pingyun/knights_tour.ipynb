```python
#chessboard=np.zeros([8,8])
#final=np.ones([8,8])
#將八個方向的步法編為list
moves=[[2,1],[1,2],[-1,2],[-2,1],[-2,-1],[-1,-2],[1,-2],[2,-1]]
#trace=[]
```

```python
#依據騎士的位置，將會超出格子的步法移除
def possible_steps():
    ns=moves.copy()
    if knight[0]==1:
        ns.remove(moves[3])
        ns.remove(moves[4])
        if knight[1]<=1:
            ns.remove(moves[5])
            ns.remove(moves[6])
            if knight[1]==0:
                ns.remove(moves[7])
        elif knight[1]>=6:
            ns.remove(moves[1])
            ns.remove(moves[2])
            if knight[1]==7:
                ns.remove(moves[0])
                
    elif knight[0]==0:
        ns.remove(moves[3])
        ns.remove(moves[4])
        ns.remove(moves[2])
        ns.remove(moves[5])
        if knight[1]<=1:
            ns.remove(moves[6])
            if knight[1]==0:
                ns.remove(moves[7])
        elif knight[1]>=6:
            ns.remove(moves[1])
            if knight[1]==7:
                ns.remove(moves[0])

    elif knight[0]==6:
        ns.remove(moves[0])
        ns.remove(moves[7])
        if knight[1]<=1:
            ns.remove(moves[5])
            ns.remove(moves[6])
            if knight[1]==0:
                ns.remove(moves[4])
        elif knight[1]>=6:
            ns.remove(moves[1])
            ns.remove(moves[2])
            if knight[1]==7:
                ns.remove(moves[3])
    elif knight[0]==7:
        ns.remove(moves[0])
        ns.remove(moves[7])
        ns.remove(moves[1])
        ns.remove(moves[6])
        if knight[1]<=1:
            ns.remove(moves[5])
            if knight[1]==0:
                ns.remove(moves[4])
        elif knight[1]>=6:
            ns.remove(moves[2])
            if knight[1]==7:
                ns.remove(moves[3])
    elif knight[1]<=1:
        ns.remove(moves[5])
        ns.remove(moves[6])
        if knight[1]==0:
            ns.remove(moves[4])
            ns.remove(moves[7])
    elif knight[1]>=6:
        ns.remove(moves[1])
        ns.remove(moves[2])
        if knight[1]==7:
            ns.remove(moves[0])
            ns.remove(moves[3])
    return ns
```

```python
#先決定騎士的初始位置，本來是完全隨機，後來決定從角落出發，增加效率
knight=[0,0]
#紀錄騎士經過的位置
trace=[]
#紀錄每個位置可能走的步法
options_list=[]
chessboard=np.zeros([8,8])
#未走滿64步時都持續搜索
while len(trace)<64:
    #如果騎士剛抵達該點，產生對應之可能步法，將步法與該位置分別append   
    if knight not in trace:
        options_list.append(possible_steps())
        trace.append(knight)
        #print('gen')
    #如果該位置仍有未走過的步法
    if options_list[-1] != []:
        #隨機選擇一步，並將該步刪除避免重複行走
        pending=random.choice(options_list[-1])
        options_list[-1].remove(pending)
        #如果走出該步後的位置沒被騎士走過
        if [knight[0]+pending[0],knight[1]+pending[1]] not in trace:
            #將騎士設至新位置
            #chessboard[knight[0],knight[1]]=1
            knight=[knight[0]+pending[0],knight[1]+pending[1]]

            #print(knight)
        #反之則回到上一步，再隨機選出新的一步
        #else:
            #print('dead')
    #當該位置沒有可走的路時(死路)
    else:
        #將該步從軌跡中刪除
        options_list.pop()
        trace.pop()
        #騎士回到上個位置
        knight=trace[-1]
        #chessboard[knight[0],knight[1]]=0
        #每次回頭時，記錄當時走到哪，是第幾步
        if len(trace) >60:
            print(knight,len(trace))
    #從上個位置還未走完的步法中，再次選擇新的一步
print(trace)
print('finished')
```
